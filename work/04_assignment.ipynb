```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('../data/state_facts.tsv', delimiter='\t')
```

```python
data.head()
```

```python
plot = data[['Area_land','Pop_2020']].plot(kind='bar',figsize=(10,10))
plot.set_xticklabels(data['State']);
plot.set_title('Land Area of 2020 Population')
```
